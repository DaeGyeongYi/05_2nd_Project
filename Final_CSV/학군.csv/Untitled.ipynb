```python
region = pd.read_csv('C:\env\Final_CSV\지역.csv\part-00000-d70adea7-1934-4300-afaa-f8ed7ff2fa29-c000.csv')
region['법정동'] = region['시도명'] + ' ' + region['법정동명']
ind=[]
for i in range(len(region)) :
    if region['시도명'][i]=='세종특별자치시' :
        ind.append('세종')
    elif len(region['시도명'][i])==4:
        ind.append(region['시도명'][i][0] +region['시도명'][i][2] +' '+ region['법정동명'][i])
    else :
        ind.append(region['시도명'][i][:2] +' '+ region['법정동명'][i])
region['short']=ind   
```

```python
region['short']
```

```python
df= pd.merge(df, region,on='법정동코드')
```
