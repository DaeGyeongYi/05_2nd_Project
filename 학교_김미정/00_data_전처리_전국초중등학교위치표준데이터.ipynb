```python
# header : 엑셀파일에서 몇번째 행을 여기로 불러올때 열제목으로 사용할 것인가 
school_data = pd.read_excel('../Data/학교_Data/1.+유·초·중등교육기관+주소록(2019).xlsx', header=3)
school_data.info()
```

```python
# drop : 필요없는 열을 제거하기 위한 함수. 다만 실제 DF에는 적용 안됨.  적용시키려면 inplace =True 사용
# columns : 이걸 사용해야 열 번호로 지정가능 아니면 한글로 쳐야됨 ['학교명', '학교상태',..]
school_data.drop(school_data.columns[[0,10,11,14,15,16]],axis=1, inplace=True)
```

```python
school_data.head()
```

```python
# 다른 테이블과 연결시키기 위해(법정동명) 도로명 주소에서 구까지 추출.
# 도로명 주소에서 스페이스 단위로 자른후 앞 두개 가져와 ind에 저장후 법정동이라는 새로운 컬럼 만듬
ind=[]
for i in range(len(school_data)):
    if school_data['도로명주소'][i].split(' ')[0]== '세종특별자치시' :
        ind.append(school_data['도로명주소'][i].split(' ')[0])
    elif school_data['도로명주소'][i].split(' ')[0]== '서울' :
         ind.append('서울특별시 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '경기' :
         ind.append('경기도 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '경남' :
         ind.append('경상남도 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '대구' :
         ind.append('대구광역시 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '충북' :
         ind.append('충청북도 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '인천' :
         ind.append('인천광역시 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '강원' :
         ind.append('강원도 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '경북' :
         ind.append('경상북도 ' +school_data['도로명주소'][i].split(' ')[1])
    elif school_data['도로명주소'][i].split(' ')[0]== '전북' :
         ind.append('전라북도 ' +school_data['도로명주소'][i].split(' ')[1])
    else :
        ind.append(' '.join(school_data['도로명주소'][i].split(' ')[:2]))

school_data['법정동']=ind
school_data
```

```python
# 외부 법정동 테이블과 연결하려면 이름을 일치시켜야 해서 이름이 다른 것들 찾아서 하나하나 지정
school_data.loc[school_data.법정동=='경남 창원시', "법정동"] = '경상남도 창원시'
school_data.loc[school_data.법정동=='경북 영천시', "법정동"] = '경상북도 영천시'
school_data.loc[school_data.법정동=='경북 영천시', "법정동"] = '경상북도 영천시'
school_data.loc[school_data.법정동=='구로구 고척로', "법정동"] = '서울특별시 구로구'
school_data.loc[school_data.법정동=='부산 사상구', "법정동"] = '부산광역시 사상구'
school_data.loc[school_data.법정동=='서울시 구로구', "법정동"] = '서울특별시 구로구'
school_data.loc[school_data.법정동=='익산시 평동로', "법정동"] = '전라북도 익산시'
school_data.loc[school_data.법정동=='인천광역시 남구', "법정동"] = '인천광역시 미추홀구'
school_data.loc[school_data.법정동=='화성시 향남읍', "법정동"] = '경기도 화성시'
school_data.loc[school_data.법정동=='경남 사천시', "법정동"] = '경상남도 사천시'
school_data.loc[school_data.법정동=='광주 광산구', "법정동"] = '광주광역시 광산구'
school_data.loc[school_data.법정동== '전남 장성군', "법정동"] = '전라남도 장성군'
```
