```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

```python
csvpath1 = '../Data/가구_Data/raw_Data/시군구_광주_전북_전남_제주_별_보험료_현황_20210915180606.csv'
csvpath2 = '../Data/가구_Data/raw_Data/시군구_대전_세종_충북_충남_별_보험료_현황_20210915180556.csv'
csvpath3 = '../Data/가구_Data/raw_Data/시군구_부산_대구_울산_경북_경남_별_보험료_현황_20210915180617.csv'
csvpath4 = '../Data/가구_Data/raw_Data/시군구_서울_인천_경기_강원_별_보험료_현황_20210915180546.csv'

tb_data1 = pd.read_csv(csvpath1, encoding='cp949')
tb_data2 = pd.read_csv(csvpath2, encoding='cp949')
tb_data3 = pd.read_csv(csvpath3, encoding='cp949')
tb_data4 = pd.read_csv(csvpath4, encoding='cp949')
```

```python
tb_data1 = tb_data1[['시군구별(1)', '시군구별(2)', '2016', '2017', '2018']]
tb_data2 = tb_data2[['시군구별(1)', '시군구별(2)', '2016', '2017', '2018']]
tb_data3 = tb_data3[['시군구별(1)', '시군구별(2)', '2016', '2017', '2018']]
tb_data4 = tb_data4[['시군구별(1)', '시군구별(2)', '2016', '2017', '2018']]
```

```python
tb_data1.drop(0, inplace = True)
tb_data1.drop(1, inplace = True)
tb_data2.drop(0, inplace = True)
tb_data2.drop(1, inplace = True)
tb_data3.drop(0, inplace = True)
tb_data3.drop(1, inplace = True)
tb_data4.drop(0, inplace = True)
tb_data4.drop(1, inplace = True)
```

```python
result = pd.concat([tb_data1, tb_data2, tb_data3, tb_data4])
```

```python
result = result.reset_index(drop=True)
```

```python
del_index_arr = []
add_df = pd.DataFrame(columns=result.columns)
for i in range(len(result)):
    second_name = result["시군구별(2)"][i]
    if(second_name == "소계" and result["시군구별(1)"][i] != "세종"):
        del_index_arr.append(i)
    elif(len(second_name.split(" ")) >= 2):
        if(len(add_df) == 0):
            add_df = add_df.append({"시군구별(1)" : result["시군구별(1)"][i],
                                    "시군구별(2)" : second_name.split(" ")[0],
                                    "2016" : result["2016"][i],
                                    "2017" : result["2017"][i],
                                    "2018" : result["2018"][i]}, ignore_index=True)
            del_index_arr.append(i)
        else:
            start = 0
            end = len(add_df)
            cnt = start
            while(True):
                if(cnt == end):
                    add_df = add_df.append({"시군구별(1)" : result["시군구별(1)"][i],
                                            "시군구별(2)" : second_name.split(" ")[0],
                                            "2016" : result["2016"][i],
                                            "2017" : result["2017"][i],
                                            "2018" : result["2018"][i]}, ignore_index=True)
                    del_index_arr.append(i)
                    break
                if(second_name.split(" ")[0] == add_df["시군구별(2)"][cnt]):
                    add_df["2016"][cnt] = str(int(add_df["2016"][cnt]) + int(result["2016"][cnt]))
                    add_df["2017"][cnt] = str(int(add_df["2017"][cnt]) + int(result["2017"][cnt]))
                    add_df["2018"][cnt] = str(int(add_df["2018"][cnt]) + int(result["2018"][cnt]))
                    del_index_arr.append(i)
                    break
                cnt = cnt+1
                
```
