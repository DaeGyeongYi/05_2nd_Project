```python
M_H_df=pd.read_csv('원본_중학교 특목고 진학정보 크롤링.csv')
M_H_df['학교명주소'].split('교')
```

```python
m=[]
n=[]
M_H_df['특목고계특목고비율']=M_H_df['특목고계특목고비율'].fillna('0')
for i in range(len(M_H_df)):
    m.append(M_H_df['학교명주소'][i].split('중학교')[0]+'중학교' +' ' + M_H_df['학교명주소'][i].split('중학교')[1].split()[0]+ ' '+M_H_df['학교명주소'][i].split('중학교')[1].split()[1])
    n.append(M_H_df['특목고계특목고비율'][i].split()[0])
```

```python
k_df =pd.DataFrame()
k_df['학교명']=m
k_df['특목고진학자수']=n
k_df
```

```python
final_m=pd.merge(k_df, M_df ,how='right', right_on='학교',left_on='학교명')
```

```python
final_m=final_m[['특목고진학자수','학교 세부 유형','학교명_y','법정동']]
```

```python
final_m=final_m.rename(columns={'학교 세부 유형':'학교종류','학교명_y':'중학교명'})
```

```python
final_m.to_csv('중학교_db.csv')
```
