```python
def create_dataframe_by_first(txt) : 
    df = pd.DataFrame()
    lis = re.split(r'[<>]', txt)    
    
    for i in range(1, len(lis), 2):
        df[lis[i].strip()] = ""
        
    return df
```

```python
endpoint = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev?'
ServiceKey = 'GuWr4NuVZZH2%2FIwkaH7Alaql4W2J7q5%2F57%2B66dUB8oiBxfAMTHBlctE64zKrxGPX3F5xsIzT96qgOEMmeP6Cxw%3D%3D'

default_pageNo = '1'
default_numOfRows = '9999'

pageNo = default_pageNo
numOfRows = default_numOfRows

for ymd_param in DEAL_YMD_arr:
    DEAL_YMD = ymd_param
    for lawd_param in LAWD_CD_arr3:
        LAWD_CD = lawd_param
        
        df_apart_price_table = pd.DataFrame(columns=df_apart_price_table.columns)
        
        paramset = "serviceKey=" + ServiceKey + "&" \
                    + "pageNo=" + pageNo + "&"  \
                    + "numOfRows=" + numOfRows + "&" \
                    + "LAWD_CD=" + LAWD_CD + "&" \
                    + "DEAL_YMD="+ DEAL_YMD
        
        url = endpoint + paramset
        
        result = requests.get(url)
        bs_obj = BeautifulSoup(result.content, 'html.parser')
        
        items = bs_obj.findAll('item')
        
        to = bs_obj.findAll('totalcount')
        try:
            if (int(to[0].text) > 9999) : 
                print(DEAL_YMD + ", " + LAWD_CD)
        except:
            pass
        for itm in items:
            dlis = re.split(r'[<>]', itm.text)
            ddic = {}
            for i in range(1, len(dlis), 2):
                ddic[dlis[i].strip()] = dlis[i+1].strip()
            df_apart_price_table = df_apart_price_table.append(ddic, ignore_index=True)
        
        df_apart_price_table.to_csv("../Data/가구_Data/raw_Data/아파트매매데이터_45750,50130/apartment_" + DEAL_YMD + "," + LAWD_CD + ".csv")
        
```
