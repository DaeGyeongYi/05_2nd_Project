## 교육청 (2016 -2020) 예산 DF 

```python
data1 = pd.read_excel('money/2017_교육청별 예산규모.xls', header=1 ) 
data1.drop(['증감액(B-A)','증감률((B-A)/A)'],axis=1, inplace =True)
data1.head()
```

```python
data2 = pd.read_excel('money/2019_교육청별 예산규모.xls', header=1) 
data2.drop(['증감액(B-A)','증감률((B-A)/A)'],axis=1, inplace =True)
data2.head()
```

```python
data3 = pd.read_excel('money/2020_교육청별 예산규모.xls', header=1) 
data3.drop(['2019년 예산(A)','증감액(B-A)','증감률((B-A)/A)'],axis=1, inplace =True)
data3.head()
```

```python
df = pd.merge(data1, data2, how = 'inner', on ='구분')
df = pd.merge(df, data3, how = 'inner', on ='구분')

df
```

## 교육청 코드 DF

```python
data = pd.read_excel('money/03.고등학교+코드_20170913.xlsx',sheet_name ='1. 고등학교 코드',usecols=[0,1,2], header=29, nrows = 17 ) 
data.columns = ['전체이름', '구분','학교코드']
data['구분'][2,3]=['경남','경북']
data['구분'][12,13]=['전남','전북']
data['구분'][15,16]=['충남','충북']
data
```

```python
code= pd.read_excel('money/기관코드 조회자료.xls',usecols=[0,1], skiprows=[2,12] ) 
code.columns = ['기관코드', '전체이름']
code
```

```python
df = pd.merge(df, data, how = 'inner', on ='구분')

df
```

```python
df = pd.merge(df, code, how = 'inner', on ='전체이름')
df
```

```python
df = df[['전체이름','구분','학교코드','기관코드','2016년 예산(A)','2017년 예산(B)','2018년 예산(A)','2019년 예산(B)','2020년 예산(B)']]
df
```

```python
df.to_csv('result/교육청.csv', encoding='utf-8', index=False)
```
