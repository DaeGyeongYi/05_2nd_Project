```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/주변환경_Data/최종본_전국지하철역목록.csv")

```

```python
df
```

```python
ea = []
```

```python
for i in df.환승노선명.values:
    try:
        if i == '-':
            ea.append(0)
        elif i =='N':
            ea.append(0)
        else:
            ea.append(len(i.split("+")))
        
    except:
        ea.append(0)
```

```python
df.환승노선명.values
```

```python
df['환승노선개수'] = ea
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.to_csv("../Data/주변환경_Data/최종본_전국지하철역목록.csv")
```
