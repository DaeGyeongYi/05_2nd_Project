```python
import pandas as pd
```

```python
d1=pd.read_csv('../학생_data/학생_2017행정구역별 학업중단자수.csv', header=0)
```

```python
ind=d1[(d1['행정구역']=='계') | (d1['학제']== '초등학교')].index
d1=d1.drop(ind)
```

```python
in1=d1[(d1['학제']== '중학교')].index
in2=d1[ d1['학제']== '고등학교'].index
in1
```

```python
d_1=d1.loc[in1]
d_1['법정동명']= d_1['시도'] + ' ' +d_1['행정구역']
d_1=d_1[['학제','학업중단자 합계','법정동명']]
d_1.rename(columns={'학업중단자 합계' :'중학교학업중단자수'}, inplace=True)
```

```python
d_2 =d1.loc[in2]
d_2['법정동명']= d_2['시도'] + ' ' +d_2['행정구역']
d_2=d_2[['학제','학업중단자 합계','법정동명']]
d_2.rename(columns={'학업중단자 합계' :'고등학교학업중단자수'}, inplace=True)
```

```python
d_drop= pd.merge(d_1,d_2, on='법정동명',how='left')
```

```python
d_drop=d_drop[['법정동명','중학교학업중단자수','고등학교학업중단자수']]
```

```python
d2=pd.read_csv('../비평준화_data/비평준화_Data_최종.csv')
i=d2[(d2['시도코드']==41)].index
d2=d2.loc[i]
d2.reset_index(drop=True,inplace=True)
```

```python
d3=pd.read_csv('../학교_data/구별_교원_진학자.csv')
i=d3[((d3['학제']=='중학교') | (d3['학제']== '고등학교'))&(d3['설립']== '소계')].index
d3=d3.loc[i]
```

```python
d3=d3[['법정동','학제','학생수_계','교원수_총계_계']]
```

```python
i=d3[(d3['학제']=='중학교')].index
d4=d3.loc[i]
i=d3[(d3['학제']=='고등학교') ].index
d5=d3.loc[i]
```

```python
d4=d4.reset_index(drop=True)
d5=d5.reset_index(drop=True)
```

```python

d5['중등학령인구'] = pd.to_numeric(d4['학생수_계'])+ pd.to_numeric(d5['학생수_계'])
d5['중등교원총계'] = pd.to_numeric(d4['교원수_총계_계'])+pd.to_numeric(d5['교원수_총계_계'])
```

```python
d5['중등학령인구']  =d5['학생수_계'].astype(int) +d5['학생수_계'].astype(int) 
d5['중등교원총계']=d4['교원수_총계_계'].astype(int)+d5['교원수_총계_계'].astype(int)
```

```python
d5=d5[['법정동','중등교원총계','중등학령인구']]
```
