## 학생-교원1인당 학생수

```python
std_per_tchr = pd.read_csv('./Data/학생_Data/교원1인당_학생수_시도_시_군_구__20210915211238.csv',encoding='cp949',header=None)
std_per_tchr = std_per_tchr.drop(index=[0,1],axis=0)
std_per_tchr.columns=["행정구역별","교원1인당 학생수","재적학생수","교원수"]
std_per_tchr
```

```python
def reanme_local():
    local_dict = dict()
    tmp=['서울특별시','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시','울산광역시',
         '세종특별자치시','경기도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주특별자치도']
    for t in tmp:
        local_dict[t]=std_per_tchr[std_per_tchr['행정구역별'] == t ].index
    return local_dict
```

## 학생-학업중단수

```python
student = pd.read_csv("./Data/학생_Data/학생_2017행정구역별 학업중단자수.csv")
```

```python
# 행정구역이 '계'인 행 
student[student['행정구역']=='계']
```

```python
# 행정구역이 소계인 행 삭제 
student.drop(student[student['행정구역']=='계'].index,inplace=True)
student
```

```python
# 행정동명 컬럼추가
student['행정동명'] = student.apply(lambda x: x['시도']+" "+ x['행정구역'],axis='columns')
```

```python
# 행정구역별 학업중단자수 합계
all_stu = student[student['학제']=='소계']
all_stu.drop(['시도','행정구역','학제'],axis=1,inplace=True)
all_stu.set_index('행정동명')
```

```python
# 초등학교별 학업중단자 합계
ele_sch = student[student['학제']=='초등학교']
ele_sch.drop(['시도','행정구역'],axis=1,inplace=True)
ele_sch.set_index('행정동명')
```

```python
# 중학교별 학업중단자 합계
mid_sch = student[student['학제']=='중학교']
mid_sch.drop(['시도','행정구역'],axis=1,inplace=True)
mid_sch.set_index('행정동명')
```

```python
# 고등학교별 학업중단자 합계
high_sch = student[student['학제']=='고등학교']
high_sch.drop(['시도','행정구역'],axis=1,inplace=True)
high_sch.set_index('행정동명')
```
